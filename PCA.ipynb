```python
import numpy as np
import pandas as pd
import pprint
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data"
names = ['sex','lenght','diameter','height','whole','shucked','viscera','shell','rings']
dataset = pd.read_csv(url,header=1, names=names)
```

```python
dataset.head()
```

```python
X = dataset.drop('sex', 1)
x = dataset.drop('rings',1)
y = dataset['rings']
```

```python
X
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=21)
```

```python
X_train.shape
```

**Data Preprocessing**

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

**Apply Random Forest classifier and get the model**

```python
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier(max_depth=8, random_state=40)
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

cm = confusion_matrix(y_test, y_pred)
print(cm)
print('Accuracy' + str(accuracy_score(y_test, y_pred)*100)[0:8]+"%")
```

**Apply Logistic Regression classifier and get the model**

```python
from sklearn.linear_model import LogisticRegression
reg_log = LogisticRegression()
reg_log.fit(X_train, y_train)
y_pred = reg_log.predict(X_test)

cm = confusion_matrix(y_test, y_pred)
print(cm)
print('Accuracy :' + str(accuracy_score(y_test, y_pred)*100)[0:3]+"%")
```

**SVC Classifier**

```python
from sklearn.svm import SVC
reg_svc = SVC()
reg_svc.fit(X_train, y_train)
y_pred = reg_svc.predict(X_test)

cm = confusion_matrix(y_test, y_pred)
print(cm)
print('Accuracy :' + str(accuracy_score(y_test, y_pred)*100)[0:10]+"%")
```

**KNN  Classifier**

```python
from sklearn.neighbors import KNeighborsClassifier
reg_knn = KNeighborsClassifier()
reg_knn.fit(X_train, y_train)
y_pred = reg_knn.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print('Accuracy' + str(accuracy_score(y_test, y_pred)))
```

# PCA

**Applying Dimensionality Reduction**

```python
print(dataset)
```

```python
from sklearn.preprocessing import StandardScaler
X_std = StandardScaler().fit_transform(X)
print (X_std[0:5])
print ("The shape of Feature Matrix is -",X_std.shape)
```

```python
X_covariance_matrix = np.cov(X_std.T)
```

```python
X_covariance_matrix
```

```python
eig_vals, eig_vecs = np.linalg.eig(X_covariance_matrix)
print('Eigenvectors \n%s' %eig_vecs) 
print('\nEigenvalues \n%s' %eig_vals)
```

```python
eig_pairs = [(np.abs(eig_vals[i]), eig_vecs[:,i]) for i in range(len(eig_vals))]
eig_pairs.sort(key=lambda x: x[0], reverse=True)
print('Eigenvalues in descending order:')
for i in eig_pairs:
    print(i[0])
```

```python
tot = sum(eig_vals)
var_exp = [(i / tot)*100 for i in sorted(eig_vals, reverse=True)]
cum_var_exp = np.cumsum(var_exp)
print ("Variance captured by each component is \n",var_exp)
print(40 * '-')
print ("Cumulative variance captured as we travel each component \n",cum_var_exp)
```

```python
print ("All Eigen Values along with Eigen Vectors")
pprint.pprint(eig_pairs)
print(40 * '-')
matrix_w = np.hstack((eig_pairs[0][1].reshape(8,1),eig_pairs[1][1].reshape(8,1)))
print ('Matrix W:\n', matrix_w)
```

```python
Y = X_std.dot(matrix_w) 
print (Y[0:5])
```

```python
Y
print(Y.shape)
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(Y, y, test_size=0.2, random_state=21)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

**Apply Random forest on the Dimension reduced data**

```python
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier(max_depth=2, random_state=0)
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)
```
